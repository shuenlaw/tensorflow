```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#函数
x = np.linspace(-1, 1, 50)
y1 = 2*x + 1
y2 = x**2
plt.plot(x, y1, x, y2)
plt.show()
```

```python
#Figure
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

#获取一张图片，可以设置图片的编号和大小等
plt.figure(num=1, figsize=(3,4))
plt.plot(x, y1)

plt.figure(num=2)
plt.plot(x, y2)
plt.plot(x, y1, 'r--', linewidth=1.0)

plt.show()
```

```python
#axes（坐标轴）
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1

plt.figure(num=1)
#设置x,y轴的取值范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))
#设置x,y轴的label
plt.xlabel("I am x")
plt.ylabel("I am y")
plt.plot(x, y1)

plt.figure(num=2)
#设置x,y轴的取值范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))
#设置x,y轴的label
plt.xlabel("I am x, new ticks")
plt.ylabel("I am y, new ticks")
#设置x轴的间隔ticks
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
#设置y轴的间隔为文字，一一对应，加上$数学字体，空格需要转义符'\ '$
#\\alpha \\beta \\gamma \\sigma \\theta
plt.yticks([-2, -1, 0, 1.5, 3],
           ['$really\ bad$', '$bad\\alpha\\beta\\gamma\\sigma\\theta$', '$normal$', '$good$', '$really\ good$'])
plt.plot(x, y1)
plt.show()
```

```python
#axes（坐标轴）,移动坐标轴位置
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=1)
plt.plot(x, y2)
plt.plot(x, y1, 'r--')

plt.figure(num=2)
plt.plot(x, y2)
plt.plot(x, y1, 'r--')
#gca = 'get current axis'
#获取图片的轴，上下左右4个(top, right, left, bottom)
ax = plt.gca()
#将right、top两轴设置为不可见
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#设置bottom轴为x轴，left轴为y轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#挪动x,y轴，选择data参数，设置到y=0,x=0
ax.spines['bottom'].set_position(('data', 0))#outward,axes
ax.spines['left'].set_position(('data', 0))
plt.show()
```

```python
#legend（图例）
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=1)
#设置label
l1, = plt.plot(x, y2, label='up')
l2, = plt.plot(x, y1, 'r--', label='down')
#设置legend
plt.legend(handles=[l1, l2,], labels=['aaa', 'bbb'], loc='best')
plt.show()
```

```python
#annotation(注解)
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1

plt.figure()
plt.plot(x, y1)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
                    
#添加值，要注释的点
x0 = 1 
y0 = 2*x0 + 1
#画点
plt.scatter(x0, y0, s=50, color='b')
#画线
plt.plot([x0,x0], [y0,0], 'k--')
#添加注释，arg1：要显示的注释,xy：显示的位置，xycoords:基于data,xytext：相对于点什么位置开始显示注释
#textcoords:注释显示基于什么，offset points 就是设置的点,fontsize:字体大小,arrowprops：箭头方向
#arrowstyle:箭头的styple,->，
plt.annotate(r'$2x+1=%s$'%y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),
            textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle='->',
            connectionstyle='arc3,rad=.2'))
#添加注释，文字,（-3.7,3）x,y轴位置
plt.text(-3.7,3,r'$This\ is\ the\ some\ annocements$', fontdict={'size':16,'color':'r'})
plt.show()
```

```python
#散点图

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
#反正切函数
#T = np.arctan2(Y,X)
T = np.random.normal(0,1,n)

plt.figure(num=1)
#s：大小，c:颜色，alpha：透明度
plt.scatter(X,Y,s=75,c=T,alpha=0.5)

plt.xlim((-1.5,1.5))
plt.ylim((-1.5,1.5))
#隐藏ticks
plt.xticks(())
plt.yticks(())
#没有设置x,y轴范围
plt.figure(num=2)
plt.scatter(X,Y,s=75,c=T,alpha=0.5)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#柱状图
n = 12
X = np.arange(n)
Y1 = np.random.normal(10,4,n)
Y2 = np.random.normal(5,5,n)

#柱状图,facecolor:主体颜色，edgecolor:边框颜色
plt.bar(X,+Y1,facecolor='r')
#plt.bar(X,-Y2,facecolor='b')
plt.xticks(np.arange(12))

#添加文字说明
for x,y in zip(X,Y1):
    #ha:横向，va:纵向
    plt.text(x,y,'%.2f'%y,ha='center',va='bottom')


plt.show()
```

```python
#等高线
def f(x,y):
    #获得z值
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256#网格密度
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
#生成网格,X(n,n)、Y(n,n)
X, Y = np.meshgrid(x,y)
#plt.scatter(X,Y)
#画等高线颜色,alpha透明度,cmap颜色地图,8等高线的层次，也就是分几个层
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.hot)
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidths=0.5)
plt.clabel(C,inline=True,fontsize=10)
plt.show()
```

```python
#3D
#导入额外的包
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
#在figure上加入3d的坐标轴
ax = Axes3D(fig)
# X,Y value
x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(x,y)
R = np.cos(X)*np.cos(Y)
#R = np.sqrt(X**2+Y**2)
# height value,z = sin(x^2 + y^2)
#Z = np.sin(R)
Z = R

# 画3d图,rstride,cstride:跨度，也就是线的密集度
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='rainbow')
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow')
ax.set_zlim(-2,2)
plt.show()
```
